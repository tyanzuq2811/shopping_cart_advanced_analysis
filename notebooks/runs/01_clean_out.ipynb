```python
# [CELL 1: PARAMETERS]
DATA_PATH = "D:/DataEngineer/DataMining/shopping_cart_advanced_analysis/data/raw/online_retail.csv"
OUTPUT_CLEANED_PATH = "data/processed/cleaned_huim_data.csv"
COUNTRY = "United Kingdom"
```

```python
import pandas as pd
df = pd.read_csv(DATA_PATH, encoding='unicode_escape')

# Loại bỏ giá trị thiếu và hóa đơn hủy (InvoiceNo bắt đầu bằng 'C')
df = df.dropna(subset=['CustomerID', 'Description'])
df = df[~df['InvoiceNo'].str.contains('C', na=False)]
df = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)]
df = df[df['Country'] == COUNTRY]

# Lưu dữ liệu sạch
import os
os.makedirs("data/processed", exist_ok=True)
df.to_csv(OUTPUT_CLEANED_PATH, index=False)
print(f"Đã làm sạch {len(df)} dòng dữ liệu. Lưu tại {OUTPUT_CLEANED_PATH}")
```
